 ##### Gráficos :
 Por otro lado, nos piden graficar los valores de $(V_{\infty} -V) / V_{\infty}$  versus $t$,  utilizando escalas lineales y escalas logarítmicas. 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bmh")

#Definir variables
list_v = [0.000, 2.718, 4.068, 4.739, 4.739, 5.237, 5.319, 5.360, 5.380, 5.390, 5.395, 5.398]
t = np.arange(0.0, 6.0, 0.5)
t2 = list(t)
```

```python
#Ciclo 
lista2 = []

for i in list_v:
   alpha = (5.4 - i)/ 5.4
   lista2 = lista2 + [alpha]
```

```python
#Grafico Escala Lineal
plt.subplot(1,2,1)
plt.plot(t2, lista2)
plt.xlabel("tiempo $[t]$")
plt.ylabel("$(V_{\infty} -V) / V_{\infty}$")
plt.errorbar(t, lista2, xerr= 0.05)
plt.title("Escala Lineal")

#Grafico Escala Logarítmica
plt.subplot(1,2,2)
plt.loglog(t2, lista2)
plt.xlabel("tiempo $[t]$")
plt.ylabel("$(V_{\infty} -V) / V_{\infty}$")
plt.errorbar(t, lista2, xerr= 0.05)
plt.title("Escala Logarítmica")

plt.subplots_adjust(wspace=0.4)
```

```python
#Grafico Eje X con Escala Log
plt.subplot(1,2,1)
plt.plot(t, lista2)
plt.xscale("log")
plt.xlabel("tiempo $[t]$")
plt.ylabel("$(V_{\infty} -V) / V_{\infty}$")
plt.errorbar(t, lista2, xerr= 0.05)
plt.title("Eje X (Log)")


#Grafico Eje Y con Escala Log
plt.subplot(1,2,2)
plt.plot(t, lista2)
plt.yscale("log")
plt.xlabel("tiempo $[t]$")
plt.ylabel("$(V_{\infty} -V) / V_{\infty}$")
plt.errorbar(t, lista2, xerr= 0.05)
plt.title("Eje Y (Log)")
plt.subplots_adjust(wspace=0.4)
```

Además nos piden graficar $ln[(V_{\infty} -V) / V_{\infty}]$ versus $t$ usando escalas lineales.

```python
beta = np.log(lista2)

plt.plot(t, beta)
plt.xlabel("tiempo $[t]$")
plt.ylabel("$ln[(V_{\infty} -V) / V_{\infty}]$")
plt.errorbar(t, beta, xerr= 0.05)
plt.title("Escala Lineal")
```
